# Natural Language Processing

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('spam_dataset.csv')
```

```python
dataset.head()
```

## Cleaning the texts

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, len(dataset)):
    mail = re.sub('[^a-zA-Z]', ' ', dataset['message_content'][i])
    mail = mail.lower()
    mail = mail.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    all_stopwords.remove('no')
    all_stopwords.remove('nor')
    mail = [ps.stem(word) for word in mail if not word in set(all_stopwords)]
    mail = ' '.join(mail)
    corpus.append(mail)
```

```python
corpus
```

## Creating the Bag of Words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, -1].values
```

```python
print(X.shape)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.80, random_state = 42)
```

## Training the Best ML model on the Training set

```python
# from sklearn.linear_model import LogisticRegression
# classifier = LogisticRegression(random_state = 42)
# classifier.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

print("\nAccuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
